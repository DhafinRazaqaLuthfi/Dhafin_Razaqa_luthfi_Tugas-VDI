# Membuat visualisasi data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# mengubah kolom 'Tanggal' menjadi tipe datetime
dataset['Tanggal'] = pd.to_datetime(dataset['Tanggal'], format='%d/%m/%Y')

# Memilih beberapa provinsi di Indonesia
Pemilihan_provinsi = ['Jawa Barat', 'DKI Jakarta', 'Sumatera Utara']

# Menentuka warna dari masing-masing Provinsi
warna = {
    'Jawa Barat': 'blue',
    'DKI Jakarta': 'green',
    'Sumatera Utara': 'red'
}

# Membuat visualisasi dengan tren harga beras
plt.figure(figsize=(12, 6))
for province in Pemilihan_provinsi:
    sns.lineplot(x=dataset['Tanggal'], y=dataset[province], label=province, color=warna[province])

# Tambahkan label serta judul pada plot
plt.title('Tren Harga Beras di Bebagai Provinsi Indonesia (2021-2024)', fontsize=16)
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Harga Beras (Rupiah)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Provinsi')

# Mengatur tata letak
plt.tight_layout()
# Menampilkan Hasil
plt.show()
```

* **Tren Naik atau Turun:** Garis yang menunjukkan kemiringan positif menunjukkan bahwa harga beras di provinsi tersebut cenderung naik selama periode yang dianalisis, sedangkan garis yang menunjukkan kemiringan negatif menunjukkan bahwa harga beras di provinsi tersebut cenderung turun.

* **Fluktuasi Harga:** Garis yang menunjukkan banyak kenaikan dan penurunan menunjukkan bahwa harga beras di provinsi tersebut telah mengalami fluktuasi besar selama periode waktu tersebut.

* **Perbandingan antarprovinsi:** Garis-garis ini menunjukkan apakah harga beras di suatu provinsi secara konsisten lebih tinggi atau lebih rendah dibandingkan provinsi lain. Misalnya, jika garis untuk DKI Jakarta berada di atas yang lain, itu menunjukkan bahwa pada periode yang sama, harga beras di DKI Jakarta lebih tinggi daripada di provinsi lain.
